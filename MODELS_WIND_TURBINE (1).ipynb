```python
def prepare_csv(df):
  # Convert the 'Time' column from string to datetime format
  df['Time'] = pd.to_datetime(df['Time'])

  # Loop through a predefined list of column names to clean and prepare the data
  for column_name in [' Wind Ø [m/s]', ' Energy prod. [kWh].1', ' Wind max. [m/s]', ' Wind min. [m/s]',
                      ' Rotation speed Ø [1/min]', ' Rotation speed max. [1/min]',
                      ' Rotation speed min. [1/min]', ' Blade angle Ø [°]']:
    # Remove any periods from the values (e.g., thousands separators)
    df[column_name] = df[column_name].str.replace('.', '', regex=False)
    # Replace commas with periods for decimal conversion
    df[column_name] = df[column_name].str.replace(',', '.')
    # Convert the column to float type for numerical processing
    df[column_name] = df[column_name].astype(float)

  # Print the data types of each column for verification and sort the DataFrame by 'Time'
  column_types = df.dtypes
  df_sorted = df.sort_values('Time')
  print(column_types)

  return df
```

```python
# Apply the prepare_csv function to the DataFrames for the years 2019–2022 to prepare the data

df = pd.read_csv('/content/merged_sorted_data_2019.csv')
df_prepared_19 = prepare_csv(df)

df = pd.read_csv('/content/merged_sorted_data_2020.csv')
df_prepared_20 = prepare_csv(df)

df = pd.read_csv('/content/merged_sorted_data_2021.csv')
df_prepared_21 = prepare_csv(df)

df = pd.read_csv('/content/merged_sorted_data_2022.csv')
df_prepared_22 = prepare_csv(df)
```

```python
def preprocessing(df):
  # Clean the column ' Energy prod. [kWh].1' by replacing non-numeric values or values >= 1000 with 0
  df[' Energy prod. [kWh].1'] = df[' Energy prod. [kWh].1'].apply(lambda x: 0 if (not isinstance(x, (int, float)) or abs(x) >= 1000) else x).fillna(0)

  print(df.info())

  # Apply linear interpolation to fill any missing values in the humidity column
  df[' Air humidity Ø [%]'] = df[' Air humidity Ø [%]'].interpolate(method='linear')

  # Calculate the mean of the 'Power Ø [kW]' column, excluding negative values
  power_mean = df[df[' Power Ø [kW]'] >= 0][' Power Ø [kW]'].mean()

  # Replace negative values in the 'Power Ø [kW]' column with the calculated mean
  df[' Power Ø [kW]'] = df[' Power Ø [kW]'].apply(lambda x: power_mean if x < 0 else x)

  # Calculate the mean values for the 'Rotation speed min. [1/min]' and 'Wind min. [m/s]' columns
  rotation_speed_min_mean = df[' Rotation speed min. [1/min]'].mean()
  wind_min_mean = df[' Wind min. [m/s]'].mean()

  # Replace values that exceed the defined thresholds with the respective mean
  df[' Rotation speed min. [1/min]'] = df[' Rotation speed min. [1/min]'].apply(lambda x: rotation_speed_min_mean if x > 28 else x)
  df[' Wind min. [m/s]'] = df[' Wind min. [m/s]'].apply(lambda x: wind_min_mean if x > 40 else x)

  # Define a list of columns to be dropped
  columns_to_drop = ['Plant', ' Power Ø [kW]', ' Rotation speed Ø [1/min]', ' Rotation speed max. [1/min]',
                     ' Rotation speed min. [1/min]', ' Ice acc. timer Ø [°Cmin]', ' Ice acc. timer Ø [°C/min]']

  # Create a new DataFrame with only the columns to be used in the models
  df_2 = df.drop(columns=columns_to_drop, axis=1)
  return df_2
```

```python
# Extract the target prediction column
target_column = ' Energy prod. [kWh].1'
y = df_2[target_column].values

# Drop the target column and other unnecessary columns for prediction from the DataFrame
X = df_2.drop([target_column, 'Time', 'month'], axis=1).values
# ' Rotation speed Ø [1/min]', ' Rotation speed max. [1/min]', ' Rotation speed min. [1/min]',
# ' Blade angle Ø [°]', ' Air humidity Ø [%]', 'Nacelle position [°]' (optional drops if not needed)

# Split the data into training and test sets (70% for training, 30% for testing)
test_percentage = 0.3
split_index = int(len(df_2) * (1 - test_percentage))
X_train = X[:split_index]
X_test = X[split_index:]
y_train = y[:split_index]
y_test = y[split_index:]

# Scale the data to the range [0, 1]
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape the data to fit the input format of the LSTM (samples, time steps, features)
X_train_reshaped = X_train_scaled.reshape(X_train_scaled.shape[0], 1, X_train_scaled.shape[1])
X_test_reshaped = X_test_scaled.reshape(X_test_scaled.shape[0], 1, X_test_scaled.shape[1])

# Create the LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
learning_rate = 0.01
optimizer = Adam(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss='mse')

# Train the model
model.fit(X_train_reshaped, y_train, epochs=50, batch_size=64, verbose=1, shuffle=False)

# Generate predictions on the test set
y_pred = model.predict(X_test_reshaped)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)
print('Mean Absolute Error:', mae)
print('R-squared (R2) Score:', r2)
print('----------')
print(y_pred)
```
